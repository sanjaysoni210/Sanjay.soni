```python
# class count
class_count_0, class_count_1 = emp_df['Attrition'].value_counts()
# Separate class
class_0 = emp_df[emp_df['Attrition'] == 0]
class_1 = emp_df[emp_df['Attrition'] == 1]# print the shape of the class
print('class 0:', class_0.shape)
print('class 1:', class_1.shape)
```

```python
class_1_over = class_1.sample(1000, replace=True)

test_over = pd.concat([class_1_over, class_0], axis=0)

print("total class of 1 and 0:",test_over['Attrition'].value_counts())

# plot the count after under-sampeling
test_over['Attrition'].value_counts().plot(kind='bar', title='count (target)')
```

```python
test_over.shape
```

```python
# Create a separate dataframe consisting only of the features i.e independent attributes
x=test_over.drop("Attrition",axis=1)
y=test_over["Attrition"]

x.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from scipy.stats import zscore
```

```python
# convert the features into z scores as we do not know what units / scales were used and store them in new dataframe
# It is always adviced to scale numeric attributes in models that calculate distances.

XScaled  = x.apply(zscore)  # convert all attributes to Z scale 

XScaled.describe()
```

```python
# Split X and y into training and test set in 75:25 ratio

X_train, X_test, y_train, y_test = train_test_split(XScaled, y, test_size=0.30, random_state=1)
```

```python
NNH = KNeighborsClassifier(n_neighbors= 5 , weights = 'distance' )
```

```python
# Call Nearest Neighbour algorithm

NNH.fit(X_train, y_train)
```

```python
# For every test data point, predict it's label based on 5 nearest neighbours in this model. The majority class will 
# be assigned to the test data point

predicted_labels = NNH.predict(X_test)
NNH.score(X_test, y_test)
```

```python
# calculate accuracy measures and confusion matrix
from sklearn import metrics

print("Confusion Matrix")
cm=metrics.confusion_matrix(y_test, predicted_labels, labels=[0, 1])

df_cm = pd.DataFrame(cm, index = [i for i in [0,1]],
                  columns = [i for i in ["Predict 0","Predict 1"]])
plt.figure(figsize = (7,5))
sns.heatmap(df_cm,annot=True,fmt='g')
```
